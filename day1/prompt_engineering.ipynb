```python
import requests
import json

def query_llama3_1(system_prompt, message):
    return query_llama(system_prompt, message, "3.1")
def query_llama3(system_prompt, message):
    return query_llama(system_prompt, message, "3")

def query_llama(system_prompt, message, model):
    messages = []
    if system_prompt:
        messages.append({
            "role": "system",
            "content": system_prompt,
        })
    messages.append({
        "role": "user",
        "content": message,
    })
    url = "http://localhost:11434/api/chat"
    data = {
        "model": f"llama{model}",
        "stream": False,
        "messages": messages
    }
    response = requests.post(url, data=json.dumps(data))
    return response.json()["message"]["content"]
```

```python
def query_openai(system_prompt, message):
    messages = []
    if system_prompt:
        messages.append({
            "role": "system",
            "content": system_prompt,
        })
    messages.append({
        "role": "user",
        "content": message,
    })
    completion = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages
    )
    output = completion.choices[0].message
    return output.content
```

```python
def call_and_print(system_prompt, prompt):
    llama3_out = query_llama3(system_prompt, prompt)
    llama3_1_out = query_llama3_1(system_prompt, prompt)
    openai_out = query_openai(system_prompt, prompt)
    print("llama3: ", llama3_out)
    print("llama3.1: ", llama3_1_out)
    print("OpenAI: ", openai_out)
```

```python
# 1.    5-shot prompt   "dog" .
query = """
Translate the following words to Korean:
1. book: 
2. cheese: 
3. water: 
4. bread: 
5. apple: 

Translate the word "dog":
"""
call_and_print("", query)
```

```python
# 2.    sentiment (positive or negative)   5-shot prompt   "The storyline was dull and uninspiring."    .
query = """
Determine the sentiment (positive or negative) of the following movie reviews:
1. "The movie was fantastic and had great acting.": positive
2. "I hated the film; it was a complete waste of time.": negative
3. "An excellent film that kept me on the edge of my seat.": positive
4. "Terrible plot and horrible acting.": negative
5. "A wonderful experience with stunning visuals.": positive

Review: "The storyline was dull and uninspiring."
Sentiment:
"""
call_and_print("", query)
```

```python
# 3.  SQL   few-shot prompt  .
query = """
Convert the following natural language requests into SQL queries:
1. "Show all employees with a salary greater than $50,000.": SELECT * FROM employees WHERE salary > 50000;
2. "List all products that are out of stock.": SELECT * FROM products WHERE stock = 0;
3. "Find the names of students who scored above 90 in math.": SELECT name FROM students WHERE math_score > 90;
4. "Retrieve the details of orders placed in the last 30 days.": SELECT * FROM orders WHERE order_date >= DATE_SUB(CURDATE(), INTERVAL 30 DAY);
5. "Get the count of customers from each city.": SELECT city, COUNT(*) FROM customers GROUP BY city;

Request: "Find the average salary of employees in the marketing department."
SQL Query:
"""
call_and_print("", query)
```

```python
query = """
# Simple - 1
Solve the following problem step-by-step: 23 + 47

Step-by-step solution:
1. Break down the numbers: 23 and 47
2. Add the tens place: 20 + 40 = 60
3. Add the ones place: 3 + 7 = 10
4. Combine the results: 60 + 10 = 70
"""
call_and_print("", query)
```

```python
query = """
Solve the following problem step-by-step: 123 - 58

Step-by-step solution:
1. Break down the numbers: 123 and 58
2. Subtract the tens place: 120 - 50 = 70
3. Subtract the ones place: 3 - 8 = -5 (borrow 10 from the tens place)
4. Adjust for borrowing: 70 - 10 + (-5) = 55
"""
call_and_print("", query)
```

```python
query = """
Solve the following problem step-by-step: 345 + 678 - 123

Step-by-step solution:
1. Break down the numbers: 345, 678, and 123
2. Add the first two numbers:
   a. 345 + 678
   b. Add the hundreds place: 300 + 600 = 900
   c. Add the tens place: 40 + 70 = 110
   d. Add the ones place: 5 + 8 = 13
   e. Combine the results: 900 + 110 + 13 = 1023
3. Subtract the third number from the result:
   a. 1023 - 123
   b. Subtract the hundreds place: 1000 - 100 = 900
   c. Subtract the tens place: 20 - 20 = 0
   d. Subtract the ones place: 3 - 3 = 0
   e. Combine the results: 900 + 0 + 0 = 900
"""
call_and_print("", query)
```

```python
query = """
Solve the following logic puzzle step-by-step:
Three friends, Alice, Bob, and Carol, have different favorite colors: red, blue, and green. We know that:
1. Alice does not like red.
2. Bob does not like blue.
3. Carol likes green.

Determine the favorite color of each friend.

Step-by-step solution:
1. Analyze the information:
   a. Carol likes green (given).
   b. Alice does not like red.
   c. Bob does not like blue.
2. Since Carol likes green, Alice and Bob cannot like green.
3. Alice does not like red, so Alice must like blue.
4. Bob does not like blue and cannot like green, so Bob must like red.
5. Verify:
   a. Carol likes green.
   b. Alice likes blue.
   c. Bob likes red.
"""
call_and_print("", query)
```

```python
query = """
Solve the following logic puzzle step-by-step:
Four people (A, B, C, D) are sitting in a row. We know that:
1. A is not next to B.
2. B is next to C.
3. C is not next to D.

Determine the possible seating arrangements.

Step-by-step solution:
1. Analyze the information:
   a. A is not next to B.
   b. B is next to C.
   c. C is not next to D.
2. List possible positions for B and C:
   a. B and C must be next to each other.
   b. Possible pairs: BC, CB.
3. Determine positions for A and D:
   a. If B is first (BCA, BCAD), then A cannot be next to B.
   b. If C is first (CAB, CABD), then D cannot be next to C.
4. Combine information to find valid arrangements:
   a. BCAD: B is next to C, A is not next to B, C is not next to D.
   b. CABD: B is next to C, A is not next to B, C is not next to D.
"""
call_and_print("", query)
```

### GPT   gpt 

```python
def query_openai_gpt(prompt, temp, max_tokens, tp, fp, pp):
    completion = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=temp,
        max_tokens=max_tokens,
        top_p=tp,
        frequency_penalty=fp,
        presence_penalty=pp,
        messages=[
                    {
                "role": "user",
                "content": prompt,
            }
        ]
    )
    output = completion.choices[0].message
    return output.content
```

```python
# default parameters
temp = 0.5
max_tokens = 1024
tp = 1
fp = 0.0
pp = 0.0

prompt = """
Tell me about a simple sentence that explains the concept of quantum mechanics.
"""
```

```python
# temperature 0, 0.3, 0.5, 1.0, 1.5, 2.0
for temp in [0, 0.3, 0.5, 1.0, 1.5, 2.0]:
    print("==== ===== Temperature: ", temp)
    print(query_openai_gpt(prompt, temp, max_tokens, tp, fp, pp))
```

```python
# top-p 0, 0.3, 0.5, 0.7, 0.9, 1.0
for tp in [0, 0.3, 0.5, 0.7, 0.9, 1.0]:
    print("==== ===== Top-P: ", tp)
    print(query_openai_gpt(prompt, temp, max_tokens, tp, fp, pp))
```
